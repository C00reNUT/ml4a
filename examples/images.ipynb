## Some image utilities

```python
from neuralarttools.util import *

# load an image from file, with an optional resize
img = load_image('../../neural-style-pt/images/inputs/monalisa.jpg', (200, 320))

# display it
display(img)

# load an image from the web, resize manually
img = load_image('https://upload.wikimedia.org/wikipedia/commons/thumb/f/f4/The_Scream.jpg/471px-The_Scream.jpg')
img = resize(img, (320, 480))
display(img)
```

### Get some info about the image

```python
w, h = get_size(img)
aspect_ratio = get_aspect_ratio(img)

print('size of last image:', w, h)
print('aspect ratio of last image:', aspect_ratio)
```

### Modify the image and save it

```python
# center square crop of the image
img = crop_to_aspect_ratio(img, 0.3)
display(img)
save(img, 'media/myImageCropped.jpg')
```
