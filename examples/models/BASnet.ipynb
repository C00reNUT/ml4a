```python
from ml4a.models import basnet
from ml4a import image, mask

img = image.load_image('https://upload.wikimedia.org/wikipedia/commons/6/6a/Mona_Lisa.jpg', (3*220, 3*350))
img_fg = basnet.get_foreground(img)
img_masked = mask.mask_image(img_fg, img)

all_images = image.concatenate_images([img, img_fg, img_masked])
image.display(all_images)
```
