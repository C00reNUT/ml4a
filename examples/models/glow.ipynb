```python
%tensorflow_version 1.x

# clone repository
!git clone https://github.com/genekogan/glow

# Get model weights and install requirements
!curl https://storage.googleapis.com/glow-demo/large3/graph_optimized.pb > graph_optimized.pb
!curl https://storage.googleapis.com/glow-demo/large3/graph_unoptimized.pb > graph_unoptimized.pb
!curl https://storage.googleapis.com/glow-demo/z_manipulate.npy > z_manipulate.npy
!curl https://storage.googleapis.com/glow-demo/shape_predictor_68_face_landmarks.dat > shape_predictor_68_face_landmarks.dat
!curl https://storage.googleapis.com/glow-demo/blocksparse-1.0.0-py2.py3-none-any.whl > blocksparse-1.0.0-py2.py3-none-any.whl
!pip install blocksparse-1.0.0-py2.py3-none-any.whl

# import libraries
import sys
sys.path.insert(0,'glow/demo')
from model import *
from google.colab import files
import cv2
from PIL import Image
import requests
from io import BytesIO
from align_face import *

# download an image from a URL
def get_face_picture(url):
  response = requests.get(url)
  img = Image.open(BytesIO(response.content))
  img = img.convert('RGB')  # if image is RGBA or Grayscale etc
  img = np.array(img)
  return img

def get_face_aligned(img):
  aligned_img, face_found = align_face(img)
  if not face_found:
    print('No face found in URL')
    return None
  return np.array(aligned_img)

def add_attribute(z, z_direction, amt):
  z_new = np.copy(z)
  z_new[0] += amt * z_direction
  return z_new

def display(img, title=None):
  if img is not None:
    pyplot.figure()
    if title is not None:
      pyplot.title(title)
    pyplot.imshow(img)
  else:
    print('No image in display')

def make_interpolation_video(z_list, num_frames_per_z, video_name):
  height, width = 256, 256
  video = cv2.VideoWriter(video_name, cv2.VideoWriter_fourcc('M','J','P','G'), fps=24, frameSize=(width,height))
  zfirst = z_list[:-1]
  zlast = z_list[1:]
  zlast.extend(z_list[0])
  z_frames = []
  for z1, z2 in zip(zfirst, zlast):
    zbetween = np.concatenate([(1.0-t) * z1 + t * z2 for t in np.linspace(0, 1, num_frames_per_z)], axis=0)
    z_frames.append(zbetween)
  z_frames = np.concatenate(z_frames, axis=0)
  n = z_frames.shape[0]
  batch_size = 16
  for f in range(0, n, batch_size):
    f1, f2 = f, min(n - 1, f + batch_size)
    batch = z_frames[f1:f2, :]
    imgs_batch = decode(batch)
    for iter in range(0,imgs_batch.shape[0]):
      video.write(imgs_batch[iter,:,:,::-1])
  cv2.destroyAllWindows()
  video.release()

tags = "5_o_Clock_Shadow Arched_Eyebrows Attractive Bags_Under_Eyes Bald Bangs Big_Lips Big_Nose Black_Hair Blond_Hair Blurry Brown_Hair Bushy_Eyebrows Chubby Double_Chin Eyeglasses Goatee Gray_Hair Heavy_Makeup High_Cheekbones Male Mouth_Slightly_Open Mustache Narrow_Eyes No_Beard Oval_Face Pale_Skin Pointy_Nose Receding_Hairline Rosy_Cheeks Sideburns Smiling Straight_Hair Wavy_Hair Wearing_Earrings Wearing_Hat Wearing_Lipstick Wearing_Necklace Wearing_Necktie Young"
tags = tags.split()
```

First we load an image of a face from a URL.

```python
#url = 'https://www.biography.com/.image/t_share/MTE4MDAzNDEwNzg5ODI4MTEw/barack-obama-12782369-1-402.jpg'
#url = 'https://i.pinimg.com/originals/0c/a4/c0/0ca4c00cb3722c07cc3b7efddfe8babe.jpg'
url = 'https://static.oprah.com/2016/01/201602-omag-wikfs-949x534.jpg'
img = get_face_picture(url)
aligned_img = get_face_aligned(img)
display(aligned_img)
```

```python
z = encode(aligned_img)
print("z is the encoding of `aligned_img`. It's shape is:", z.shape)

decoded_img = decode(z)   # z can be a list of encodings
display(decoded_img[0])
```

Now let's do some "operations".

By moving in z-space, we can make changes to the encoded image. Because z is such a high-dimensional space, most directions won't do anything very meaningful. 

For instance, here is adding a random vector.


```python
z_add = np.random.normal(0, 1, z.shape[-1])

# this adds 0.3 * z_add to the Obama z vector
z_new = add_attribute(z, z_add, 0.3)

# # let's see how the image changes progressively
z_new = [add_attribute(z, z_add, amt) for amt in [0, 0.25, 0.5, 1.0]]
imgs_new = [decode(z_)[0] for z_ in z_new]

display(np.concatenate(imgs_new, axis=1))
```

In order to find a vector of an attribute which is interpretable, a generic strategy is to find many z-vectors of images which have that attribute, and many z-vectors of images which don't have that attribute, average each group together, and take the vector between them. 

Since this strategy is tedious, OpenAI found a bunch of them for us. The z-vectors for these attributes can be found in the map variable `z_manipulate`.

### Known attribute vectors

`5_o_Clock_Shadow`, `Arched_Eyebrows`, `Attractive`, `Bags_Under_Eyes`, 
`Bald`, `Bangs`, `Big_Lips`, `Big_Nose`, `Black_Hair`, `Blond_Hair`, 
`Blurry`, `Brown_Hair`, `Bushy_Eyebrows`, `Chubby`, `Double_Chin`, 
`Eyeglasses`, `Goatee`, `Gray_Hair`, `Heavy_Makeup`, `High_Cheekbones`, 
`Male`, `Mouth_Slightly_Open`, `Mustache`, `Narrow_Eyes`, `No_Beard`, 
`Oval_Face`, `Pale_Skin`, `Pointy_Nose`, `Receding_Hairline`, `Rosy_Cheeks`, 
`Sideburns`, `Smiling`, `Straight_Hair`, `Wavy_Hair`, `Wearing_Earrings`, 
`Wearing_Hat`, `Wearing_Lipstick`, `Wearing_Necklace`, `Wearing_Necktie` `Young`


```python

# Blond hair
z_blond = add_attribute(z, z_manipulate[tags.index('Blond_Hair')], 1.25)
img_blond = decode(z_blond)
display(img_blond[0], 'add 1.25 x Blond')

# Goatee
z_goatee = add_attribute(z, z_manipulate[tags.index('Goatee')], 1.25)
img_goatee = decode(z_goatee)
display(img_goatee[0], 'add 1.25 x Goatee')

# Female (-male)
z_female = add_attribute(z, z_manipulate[tags.index('Male')], -1.25)
img_female = decode(z_female)
display(img_female[0], 'subtract 1.25 x Male')

# Smiling
z_smiling = add_attribute(z, z_manipulate[tags.index('Smiling')], 1.25)
img_smiling = decode(z_smiling)
display(img_smiling[0], 'add 1.25 x Smiling')

# Eyeglasses
z_eyeglasses = add_attribute(z, z_manipulate[tags.index('Eyeglasses')], 1.25)
img_eyeglasses = decode(z_eyeglasses)
display(img_eyeglasses[0], 'add 1.25 x Eyeglasses')

# Old (-young)   ** note, OpenAI's dataset is mislabeled... `Young` is actually its opposite: an old vector
z_young = add_attribute(z, z_manipulate[tags.index('Young')], -1.25)
img_young = decode(z_young)
display(img_young[0], 'subtract 1.25 x Young (mislabeled)')

# Young
z_old = add_attribute(z, z_manipulate[tags.index('Young')], 1.25)
img_old = decode(z_old)
display(img_old[0], 'add 1.25 x Young (mislabeled)')
```

You can make an interpolation video through a list of z_vectors with `num_frames_per_z` for each transition (looping around to the beginning). This may take a while if you are making many frames. Try starting small.

```python
video_name = 'operah_interpolations.mov'
z_list = [z_blond, z_goatee, z_female, z_smiling, z_eyeglasses, z_young, z_old]
num_frames_per_z = 30

make_interpolation_video(z_list, num_frames_per_z, video_name)
```
