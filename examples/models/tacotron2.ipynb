# Run tacotron2

```python
from ml4a import audio
from ml4a.models import tacotron2

text = 'Hello everyone!'
speech = tacotron2.run(text)
```

`speech` contains the raw waveform and sampling rate, which can be played back.

```python
audio.display(speech.wav, speech.sampling_rate)
```

You can also plot the waveform.

```python
%matplotlib inline
audio.plot(speech.wav, speech.sampling_rate)
```
